```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
```

```python
frequency = 800
samples = np.arange(0, 100)
x = samples / 44100
y =  1 * np.sin(2 * np.pi * frequency * x)
rate = 1000
z = 1 - np.exp(-rate*x)
k = np.exp(-10*x)
plt.plot(x,y)
plt.plot(x,z)
plt.plot(x,k)
plt.plot(x,y*z)
plt.plot(x,y*k)
plt.show()
```

```python
x = np.arange(100) / 44100
y = np.sin(2 * np.pi * frequency * samples)
k = np.exp(-frequency*1.5*x)
plt.plot(samples,y)
plt.plot(samples, y*k)
```
