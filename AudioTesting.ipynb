```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
sampleRate = 44100
frequency = 2000
duration = 1/frequency
x = np.linspace(0, duration, int(sampleRate*duration))
y =  0.1 * np.sin(2 * np.pi * frequency * x)
plt.plot(x,y)
```

```python
samples = np.arange(100, 100 + 200)
x = samples / 44100
y =  0.1 * np.sin(2 * np.pi * frequency * x)
plt.plot(x,y)
```
